```python
import random
import time
%run pravljenjeGrafova.ipynb

def najveciStepen(graf):      
    return max(graf, key=lambda key: len(graf[key]))
       
def najveciStepen2(graf,lista):   
    if(len(lista)>2):     
       return max(lista, key=lambda key: nadjiBrSusjeda(graf,key))
    else:
        return lista[0]

def nadjiSusjede (graf,cvor):
    return graf[cvor].copy()

def nadjiBrSusjeda (graf,cvor):
    return len(graf[cvor])

#Trazi da li je datom cvoru ijedan susjed dom cvor
def je_dom(cvor, domSkup,graf):
    susjedi = nadjiSusjede(graf,cvor)
    return any(susjed in domSkup for susjed in susjedi)
```

```python
#Glavna funkcija

def nadji_MSPDS(graf):
    
    domSkup = set()
    oznaceni = set()
    zaObici = []

    trcvor = najveciStepen(graf)

    domSkup.add(trcvor)
    oznaceni.add(trcvor)
    zaObici = nadjiSusjede(graf,trcvor)

    for l in zaObici:
            if l not in oznaceni:
                oznaceni.add(l)  
    
    while len(zaObici)>0:

        br = najveciStepen2(graf,zaObici)       # Biramo najveci stepen cvora koji se nalazi u listi zaObici
        trcvor = br
        zaObici.remove(br)

        lista1 = nadjiSusjede(graf,trcvor)

        for l in lista1:
            if  l not in oznaceni:
                zaObici.append(l)
            oznaceni.add(l)
         
        if not je_dom(trcvor,domSkup,graf):
            oznaceni.add(trcvor)
            domSkup.add(trcvor)
  
    return domSkup
```
