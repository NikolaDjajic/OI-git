```python
import random
import time

def najveciStepen(graf):      
    return max(graf, key=lambda key: len(graf[key]))
       
def najveciStepen2(graf,lista):   
    if(len(lista)>2):     
       return max(lista, key=lambda key: nadjiBrSusjeda(graf,key))
    else:
        return lista[0]

def nadjiSusjede (graf,cvor):
    return graf[cvor]

def nadjiBrSusjeda (graf,cvor):
    return len(graf[cvor])

#Trazi da li je datom cvoru ijedan susjed dom cvor
def je_dom(cvor, domSkup,graf):
    susjedi = nadjiSusjede(graf,cvor)
    return any(susjed in domSkup for susjed in susjedi)
```

```python
#Glavna funkcija

def nadji_MSPDS(graf1):
    

    domSkup = set()
    oznaceni = set()
    zaObici = []

    #Uzima prvi nasumicno
    #trcvor = najvise(graf,[])

    #Uzima prvi po stepenu cvora
    trcvor = najveciStepen(graf1)
   # print("Pocetni cvor:",trcvor)

    domSkup.add(trcvor)
   # print("DODAJEM U DOM SKUP",trcvor)

    oznaceni.add(trcvor)
    zaObici = nadjiSusjede(graf1,trcvor)

    for l in zaObici:
            if l not in oznaceni:
                oznaceni.add(l)  

    
   # print("Za obici:",zaObici)
  #  print("Oznaceni:",oznaceni)
   # print("-----------------------------------------")
    
    while len(zaObici)>0:
        graf=graf1
        #br = random.randint(0,len(zaObici)-1)    Random
        br = najveciStepen2(graf,zaObici)       # Biramo najveci stepen cvora koji se nalazi u listi zaObici

        #print(br)
        trcvor = br
        zaObici.remove(br)

        lista1 = nadjiSusjede(graf,trcvor)

        for l in lista1:
            if  l not in oznaceni:
                zaObici.append(l)
            oznaceni.add(l)
        
     #   print("Trenutni cvor",trcvor)
     #   print("Za obici",zaObici)
     #   print("Oznaceni",oznaceni)
        
        if not je_dom(trcvor,domSkup,graf):
            oznaceni.add(trcvor)
            domSkup.add(trcvor)
        #    print("DODAJEM U DOM SKUP",trcvor)
       # print("-----------------------------------------")
    return domSkup
```
